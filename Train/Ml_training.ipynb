```python
def read_astro_column_name(file_path):
    # Open and read the CSV file content to extract the header (first line)
    with open(file_path, "r", encoding="utf-8") as file:
        first_line = file.readline().strip()

    # Split the first line by commas and clean any extra spaces
    column_names = [name.strip() for name in first_line.split(",")]

    return column_names

# Adjust the file path to your dataset location
file_path = "C:/Users/USER/Documents/GitHub/Infinite_Stratos_Project/data/xxx_dataset.csv"

# Read the column names from the dataset
column_names = read_astro_column_name(file_path)

# Load the CSV file into a DataFrame using the correct column names
df = pd.read_csv(
    file_path,
    header=None,  # Since we're specifying the column names manually
    names=column_names,
    sep=",",  # Assuming it's comma-separated
    engine="python"
)

# Display the first few rows to verify everything works as expected
print(df.head())
```

```python
for column in df.columns:
    j = df[column].value_counts(dropna=False)
    print(column, ":", j)
    print("----------------------------------------")
```

```python
for column in df.columns:
    df.drop(df[df[column] == "?"].index, inplace=True)
df.columns
```

```python
def map_data(x):
    convert = x.unique()
    return x.map(dict(zip(convert, range(1, len(convert) + 1))))


for column in df.columns:
    if df[column].dtype == "object":
        df[column] = map_data(df[column])
df.astype(int)
```

```python
df_test = pd.read_csv(
    "C:/Users/USER/Documents/GitHub/Infinite_Stratos_Project/data/xxx_dataset.csv",
    header=None,
    names=column_names,
    sep=",",  # Use just a comma as separator
    engine="python",
)
```

```python
for column in df_test.columns:
    df_test.drop(df_test[df_test[column] == "?"].index, inplace=True)

for column in df_test.columns:
    if df_test[column].dtype == "object":
        df_test[column] = map_data(df_test[column])
df_test.astype(int)
```

```python
# Clean up the column names
df.columns = df.columns.str.replace('"', '')
df_test.columns = df_test.columns.str.replace('"', '')

# Drop rows with missing 'u' values (if necessary)
df = df.dropna(subset=['Temperature'])
df_test = df_test.dropna(subset=['Temperature'])

# Split into features and target
X_train = df.drop('Temperature', axis=1)
y_train = df['Temperature']

X_test = df_test.drop('Temperature', axis=1)
y_test = df_test['Temperature']
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
import joblib

# ✅ Reduce training set size to 100 samples
X_train = X_train.sample(n=241, random_state=42)
y_train = y_train.loc[X_train.index]

# ✅ Scale the sampled data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Corrected variable name
X_test_scaled = scaler.transform(X_test)
print(X_train_scaled.shape)
print(y_train.shape)

# ✅ Define hyperparameters for SVM model
param_grid = {
    'C': [0.1, 1, 10],  
    'gamma': ['scale', 0.01, 0.1],  
    'kernel': ['rbf', 'poly'],  
    'degree': [2, 3]  # Only for 'poly' kernel  
}

# ✅ Initialize SVM with balanced class weights
svm = SVC(kernel="linear", class_weight="balanced", probability=True, random_state=42)

# ✅ Perform hyperparameter tuning manually
best_svm = None
best_score = -1

# Try different hyperparameters manually and pick the best one
for C in param_grid['C']:
    for gamma in param_grid['gamma']:
        svm.set_params(C=C, gamma=gamma)
        svm.fit(X_train_scaled, y_train)
        
        # Evaluate the model
        y_pred = svm.predict(X_test_scaled)
        accuracy = accuracy_score(y_test, y_pred)
        
        # Update the best model if we get a higher score
        if accuracy > best_score:
            best_score = accuracy
            best_svm = svm

# ✅ Best hyperparameters
print(f"Best Hyperparameters: C={best_svm.C}, gamma={best_svm.gamma}")

# ✅ Evaluate the best model
y_pred = best_svm.predict(X_test_scaled)

# ✅ Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Test accuracy: {accuracy:.2f} %")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# ✅ Save the trained model and scaler
joblib.dump(best_svm, "C:/Users/USER/Documents/GitHub/Infinite_Stratos_Project/exported_model/rf/svm/svm_Budget_model.pkl")
joblib.dump(scaler, "C:/Users/USER/Documents/GitHub/Infinite_Stratos_Project/exported_model/rf/svm/svm_Budget_scaler.pkl")
```
